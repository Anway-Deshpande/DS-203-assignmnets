```python
# importing libraries

import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

**Reading the file and displaying the information**

```python
data_file = pd.read_excel('NDAP_REPORT_7004.xlsx')

display(data_file)
```

**Printing number of coulmns and rows in the file**

```python
print('number of rows: ',str(data_file.shape[0]))
print('number of columns: ',str(data_file.shape[1]))
```

Printing unique values in each column

```python
for col in data_file:
    unq_val = pd.unique(data_file[col])
    print(col + ' has ' + str(len(unq_val)) + ' values')
```

Printing data type of columns

```python
data_file.dtypes
```

```python
sum_nan = []
for col in data_file:
    if data_file.columns.get_loc(col) >= 9:
        sum_nan.append(data_file[col].isna().sum())
plt.hist(sum_nan)
```

```python
for col in data_file:
    if data_file.columns.get_loc(col) >= 9 :
        data_file[col] = data_file[col].replace(['zero'],0)
        data_file[col] = data_file[col].replace([np.nan],0)
        data_file[col] = data_file[col].astype(int)
display(data_file)
```
